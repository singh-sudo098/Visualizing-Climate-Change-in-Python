# Editing csv data file

```python
temp_df = pd.read_csv(dataset_path + "EDITED_temperature.csv")
temp_df.head()
```

```python
temp_df.info()
```

```python
temp_df.Year = pd.to_datetime(temp_df['Year'])
temp_df.Temperature = temp_df.Temperature.astype(float)
temp_df.Temperature_normalised = temp_df.Temperature_normalised.astype(float)
temp_df.geometry = temp_df.geometry.astype(object)
temp_df.info()
```

```python
temp_df.info()
```

```python
temp_df.head()
```

# Editing the date in the dataframe

```python
from datetime import date
from itertools import product

y = np.arange(1990,2020)
d = [15]
m = [1]
cols = ['Year','Month','Day']
new_year_list = pd.DataFrame(list(product(y, m, d)), columns=cols)

DATE = []
for y, m in zip(new_year_list.Year, new_year_list.Month):
    DATE.append(date(y, 1, 15))


new_year_list['Date'] = DATE
new_year_list.info()
```

```python
new_year_list.head(30)
```

```python
new_year_list.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
new_year_list.head()
```

```python
country_list = pd.read_csv(dataset_path + "country_list.csv")
country_list.head()
```

```python
country_list.info()
```

```python
def cartesian_product_simplified(left, right):
    la, lb = len(left), len(right)
    ia2, ib2 = np.broadcast_arrays(*np.ogrid[:la,:lb])

    return pd.DataFrame(
        np.column_stack([left.values[ia2.ravel()], right.values[ib2.ravel()]]))
```

```python
new_temp_df = cartesian_product_simplified(country_list, new_year_list)
new_temp_df.head(34)
```

```python
new_temp_df.info()
```

```python
new_temp_df.rename(columns = {0:'Country_2',
                           1 : 'Date'
                          }, 
                inplace = True)
new_temp_df.head()
```

```python
new_temp_df.Date = pd.to_datetime(new_temp_df.Date)
new_temp_df.info()
```

```python
temp_df2 = pd.concat([temp_df, new_temp_df], axis=1, join="inner")
temp_df2.head(34)
```

```python
temp_df2.info()
```

```python
temp_df2.columns
```

```python
# Not taking the geometry column because null geometry values will not be filled automatically with ffill method
temp_df2 = temp_df2[['Country', 'Date', 'Temperature', 'Temperature_normalised']]
temp_df2.head()
```

# Creating new date dataframe for dates

```python
from datetime import date
from itertools import product

y = np.arange(1990,2020)
d = [15]
m = np.arange(1,13)
cols = ['Year','Month','Day']
month_year_df = pd.DataFrame(list(product(y, m, d)), columns=cols)

DATE = []
for y, m in zip(month_year_df.Year, month_year_df.Month):
    DATE.append(date(y, m, 15))


month_year_df['Date'] = DATE
month_year_df.info()
```

```python
month_year_df.head()
```

```python
month_year_df.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
month_year_df.head()
```

```python
country_date_df = cartesian_product_simplified(country_list, month_year_df)
country_date_df.head(34)
```

```python
country_date_df.info()
```

```python
country_date_df.rename(columns = {0:'Country',
                                  1 : 'Date'
                                 }, 
                       inplace = True)
country_date_df.head()
```

```python
country_date_df.Date = pd.to_datetime(country_date_df.Date)
country_date_df.info()
```

# Merging the dataframes to form a monthly dataset

```python
merged = pd.merge(country_date_df, temp_df2, how='left', left_on=['Country', 'Date'], right_on=['Country', 'Date'])
merged.head(34)
```

```python
merged = merged.fillna(method='ffill')
merged.head(34)
```

```python
merged.info()
```

# Making a geometry dataframe for the merged dataset and then joining them

```python
map_df = pd.read_csv(dataset_path + 'geopandas_mapping.csv')
map_df.head()
```

```python
map_df.info()
```

```python
map_df.geometry = map_df.geometry.apply(wkt.loads)
geometry = map_df.geometry
crs = {'init': 'epsg:4326'}
map_df = gpd.GeoDataFrame(map_df, crs=crs, geometry=geometry)
```

```python
long_country_list_df = merged[['Country']]
long_country_list_df.info()
```

```python
geometry_df = pd.merge(long_country_list_df, map_df, on='Country')
# geometry_country_df = cartesian_product_simplified(long_country_list_df, geometry_df)
geometry_df.info()
```

```python
geometry_df.rename(columns = {'Country':'Country_2'
                             }, 
                   inplace = True)
```

```python
geometry_df.head()
```

# Final dataset by concatenating the geometry_df with the merged_df

```python
final = pd.concat([merged, geometry_df], axis=1, join="inner")
final.info()
```

```python
final.columns
```

```python
final = final[['Country', 'Date', 'Temperature', 'Temperature_normalised', 'geometry']]
```

```python
edited_master = final
edited_master = edited_master.astype(object)
edited_master.to_csv(dataset_path + 'FINAL_TEMPERATURE.csv', encoding='utf-8', index=False)
```
