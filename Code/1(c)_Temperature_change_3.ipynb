```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import time
import geopandas as gpd
from shapely import wkt
```

# Editing csv data file

```python
temp_df = pd.read_csv(dataset_path + "EDITED_temperature.csv")
temp_df.head()
```

```python
temp_df.info()
```

```python
temp_df.Year = pd.to_datetime(temp_df['Year'])
temp_df.Temperature = temp_df.Temperature.astype(float)
temp_df.Temperature_normalised = temp_df.Temperature_normalised.astype(float)
temp_df.geometry = temp_df.geometry.astype(object)
temp_df.info()
```

```python
temp_df.geometry = temp_df.geometry.apply(wkt.loads)
geometry = temp_df.geometry
crs = {'init': 'epsg:4326'}
temp_df = gpd.GeoDataFrame(temp_df, crs=crs, geometry=geometry)
```

```python
temp_df.info()
```

```python
temp_df.head()
```

# Editing the date in the dataframe

```python
from datetime import date
from itertools import product

y = np.arange(1990,2020)
d = [15]
m = [1]
cols = ['Year','Month','Day']
new_year_list = pd.DataFrame(list(product(y, m, d)), columns=cols)

DATE = []
for y, m in zip(new_year_list.Year, new_year_list.Month):
    DATE.append(date(y, 1, 15))


new_year_list['Date'] = DATE
new_year_list.info()
```

```python
new_year_list.head(30)
```

```python
new_year_list.drop(['Year', 'Month', 'Day'], axis=1, inplace=True)
new_year_list.head()
```

```python
country_list = pd.read_csv(dataset_path + "country_list.csv")
country_list.head()
```

```python
country_list.info()
```

```python
def cartesian_product_simplified(left, right):
    la, lb = len(left), len(right)
    ia2, ib2 = np.broadcast_arrays(*np.ogrid[:la,:lb])

    return pd.DataFrame(
        np.column_stack([left.values[ia2.ravel()], right.values[ib2.ravel()]]))
```

```python
new_temp_df = cartesian_product_simplified(country_list, new_year_list)
new_temp_df.head(34)
```

```python
new_temp_df.info()
```

```python
new_temp_df.rename(columns = {0:'Country_2',
                           1 : 'Date'
                          }, 
                inplace = True)
new_temp_df.head()
```

```python
new_temp_df.Date = pd.to_datetime(new_temp_df.Date)
new_temp_df.info()
```

```python
temp_df2 = pd.concat([temp_df, new_temp_df], axis=1, join="inner")
temp_df2.head(34)
```

```python
temp_df2.info()
```

```python
temp_df2.columns
```

```python
# Not taking the geometry column because null geometry values will not be filled automatically with ffill method
temp_df2 = temp_df2[['Country', 'Date', 'Temperature', 'Temperature_normalised']]
temp_df2.head()
```
